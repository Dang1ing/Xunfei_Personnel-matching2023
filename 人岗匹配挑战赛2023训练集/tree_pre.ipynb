```python
import pandas as pd
import sklearn 
from sklearn import preprocessing
df=pd.read_csv("raw.csv",header=0)
```

```python

print(df["工作经历数量"].max())
print(df["社会经历数量"].max())
print(df["项目数量"].max())
```

```python
#labelencoder
pos_set=set()
for pos in df["工作职位"]:
    if pos !='[]':
        res=pos.strip('[')
        res=res.strip(']')
        res=res.split(',')
        for p in res:
            p=p.strip()
            pos_set.add(p)

pos_set=list(pos_set)
pos_set
```

```python
pos_enc=preprocessing.LabelEncoder()
pos_enc=pos_enc.fit(pos_set)
for index,row in df.iterrows():
    p_lis=list()
    if row["工作职位"]!="[]":
        res=row["工作职位"].strip('[')
        res=res.strip(']')
        res=res.split(',')
        for p in res:
            p=p.strip()
            p_lis.append(p)
        data=pos_enc.transform(p_lis)
        row["工作职位"]=data
        df.iloc[index]=row
```

```python
#仿照上述做法，将剩余特征进行编码
lis=["公司名称","专业","学校","学校层次","职位名","组织名","项目名称","期望职位"]
for l in lis:
    pos_set=set()
    for pos in df[l]:
        if pos !='[]':
            res=pos.strip('[')
            res=res.strip(']')
            res=res.split(',')
            for p in res:
                p=p.strip()
                pos_set.add(p)

    pos_set=list(pos_set)

    pos_enc=preprocessing.LabelEncoder()
    pos_enc=pos_enc.fit(pos_set)
    for index,row in df.iterrows():
        p_lis=list()
        if row[l]!="[]":
            res=row[l].strip('[')
            res=res.strip(']')
            res=res.split(',')
            for p in res:
                p=p.strip()
                p_lis.append(p)
            data=pos_enc.transform(p_lis)
            row[l]=data
            df.iloc[index]=row

df
```
